```python
# -----------
# IDev.
# Coding Tips
# Copyright Â© IDev_IO. All rights reserved
# -----------

# ðŸ“Œ Mysql Fake Database Seeder

# Used Packages:
# ðŸ“Œ prettytable      To make data pretty and readable when i show it to users. 
# ðŸ“Œ faker            To seed fake data in database.
# ðŸ“Œ mysql.connector  To create mysql database connection.


from prettytable import PrettyTable
from faker import *              
import mysql.connector
from mysql.connector import connect


# Create mysql connection to DB
user_name = 'Your_Database_User_Name'
user_password = 'Your_Database_Password'
host_name = 'Your_Database_host_Name'
db_name = 'Your_Database_Name'

db = connect(user = user_name, 
             password = user_password, 
             host = host_name, 
             database = db_name)

cursor = db.cursor(buffered = True)


# Create Database table 

create_query = """CREATE TABLE developers( 
            id int(255) not null auto_increment primary key, 
            name varchar(255), 
            job varchar(255),
            phoneNO varchar(255),
            address varchar(255),
            bio text(255)
            );"""

cursor.execute(create_query)
```

```python
fake = Faker('en_US')

# Create function to execute mysql queries
def execute_queries(query):
    try:
        cursor.execute(query)
        db.commit()
    except:
        db.rollback()

 
for _ in range(0, 100):
    
    dev_name = fake.name()
    dev_job = fake.job()
    dev_phoneNO = fake.phone_number()
    dev_address = fake.address()
    dev_bio = fake.paragraph(nb_sentences=3)
    
    insert_query = "INSERT INTO developers(name, job, phoneNO, address, bio) VALUES ('%s', '%s', '%s', '%s', '%s')"\
                    %  (dev_name, dev_job, dev_phoneNO, dev_address, dev_bio)

    execute_queries(insert_query)
```

```python
def get_data(query):
    try:
        cursor.execute(query)
        data = cursor.fetchall()
        
        fields = PrettyTable()
        fields.field_names = ["ID", "Developer Name", "Job", "Phone No."]

        for row in data:
            fields.add_row([row[0], row[1], row[2], row[3]])
        
        print(fields)
            
    except:
        db.rollback()
    finally:
        db.close()


read_query = "SELECT * FROM developers"

get_data(read_query)
```
